```python
import math
import json
import re
import collections
from os import path
import pandas as pd
import numpy as np
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import AlignIO
from Bio.Align import MultipleSeqAlignment
from Bio.Align import AlignInfo
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import seaborn as sns
```

```python
def simpsons_index(seq_list, num_codons, kind):
    """
    D = 1 - summation(pi^2)
    pi = relative frequency of the ith codon at the residue
    """
    
    #list of lists of all codons (or aas) seen at a given position. 
    #len list of lists= num_codons. len of lists= num isolates (that were sequenced at this position)  
    all_codons_at_all_positions = []
    
    #for each codon/aa position
    for i in range(num_codons):
        codons_at_pos_i = []
        #for each isolate sequence
        for x in seq_list:
            #only consider codons or aas that are fully sequenced (no ambiguous bases)
            if kind == 'codon':
                match = re.search(r"[ATGC]{3}", str(x[i]))
            elif kind == 'aa':
                match = re.search(r"[ARNDCQEGHILKMFPSTWYV]{1}", str(x[i]))
            if match:
                codons_at_pos_i.append(x[i])

        all_codons_at_all_positions.append(codons_at_pos_i)
    
    #list of simpsons D by residue position
    simpsons_d_by_position = []
    
    #count codons seen at each position
    for i in range(num_codons):
        list_of_codons = all_codons_at_all_positions[i]
        codon_counter = collections.Counter(list_of_codons)
        codon_counts = list(codon_counter.values())
        codon_freqs = [c/sum(codon_counts) for c in codon_counts]
        if len(codon_freqs) != 0:
            simpsons_d = 1.0 - sum([freq**2 for freq in codon_freqs])
        simpsons_d_by_position.append(simpsons_d)
        
    
    return simpsons_d_by_position
```

```python
def compute_diversity(cov, gene):
    
    input_file_alignment = '../'+str(cov)+'/results/aligned_'+str(cov)+'_'+str(gene)+'.fasta'
    input_reference = f"../{cov}/config/{cov}_{gene}_reference.gb"

    with open(input_reference, "r") as reference_handle:
        for record in SeqIO.parse(reference_handle, "genbank"):
            gene_len = len(record.seq)
            num_codons = int(gene_len/3)
    
    aa_list = []
    codon_list = []
    
    with open(input_file_alignment, "r") as aligned_handle:
        for virus in SeqIO.parse(aligned_handle, "fasta"):
            virus_aa = virus.seq.translate()
            aa_list.append(str(virus_aa))
            virus_codons = [str(virus.seq)[i:i+3] for i in range(0, gene_len, 3)]
            codon_list.append(virus_codons)
            
    simpsons_d_codon = simpsons_index(codon_list, num_codons, 'codon')
    simpsons_d_aa = simpsons_index(aa_list, num_codons, 'aa')
    
    return simpsons_d_codon, simpsons_d_aa
```

```python
def plot_diversity_comparison(cov, genes):
    
    
    data_to_plot = []
    for gene in genes:
        simpsons_d_codon, simpsons_d_aa = compute_diversity(cov, gene)
        if len(simpsons_d_codon)==len(simpsons_d_aa):
            for i in range(len(simpsons_d_codon)):
                codon_d = simpsons_d_codon[i]
                aa_d = simpsons_d_aa[i]
                #determine if point falls on line xy or x=o
                #draw cut-off at 0.05 because hard to tell below that
                relationship= 'unknown'
                
                if aa_d >= 0.05:
                    if codon_d+0.03 > aa_d > codon_d-0.03:
                        relationship = 'xy'
                
                elif codon_d >= 0.05:
                    if aa_d <= 0.03:
                        relationship = 'x0'


                data_to_plot.append({'codon_d': codon_d, 'aa_d': aa_d, 
                                     'gene': gene, 'pos': i, 'relationship': relationship})
        else:
            print('codon and aa lists are not of equal length')
    
    df_to_plot = pd.DataFrame(data_to_plot)
    

    #count
    fraction_xy = {}
    fraction_x0 = {}
    
    for gene in genes:
        gene_len = len(df_to_plot[df_to_plot['gene']==gene])
        gene_xy = len(df_to_plot[(df_to_plot['gene']==gene) & (df_to_plot['relationship']=='xy')])
        fraction_xy[gene] = gene_xy/gene_len
        gene_x0 = len(df_to_plot[(df_to_plot['gene']==gene) & (df_to_plot['relationship']=='x0')])
        fraction_x0[gene] = gene_x0/gene_len

        

    #plot amino acid versus codon diversity
    color_map = {'rdrp':'#5c5c5c', 'spike': '#ff7400', 's1': '#ff9a00', 's2': '#858585'}
    color_map_relationship = {'xy': '#76C7BE', 'x0': '#0B194C', 'unknown': '#858585'}
    
    fig, ax = plt.subplots(figsize=(8,8))
    fig = sns.scatterplot(data=df_to_plot, x= 'codon_d', y = 'aa_d', color = '#858585')

    #color by relationship
    fig1, ax1 = plt.subplots(figsize=(8,8))
    fig1 = sns.scatterplot(data=df_to_plot, x= 'codon_d', y = 'aa_d', hue = 'relationship', palette = color_map_relationship)

    #color by gene
    fig2, ax2 = plt.subplots(figsize=(8,8))
    fig2 = sns.scatterplot(data=df_to_plot, x= 'codon_d', y = 'aa_d', hue = 'gene', palette = color_map)
    
    for axis in [ax, ax1, ax2]:
        axis.set_xlabel('Codon diversity')
        axis.set_ylabel('Amino Acid diversity')
    
    relationship_count_text = f"Fraction of residues along x=y:\n  {genes[0]}:{fraction_xy[genes[0]]}\n  {genes[1]}:{fraction_xy[genes[1]]}\n\nFraction of residues along x=0:\n  {genes[0]}:{fraction_x0[genes[0]]}\n  {genes[1]}:{fraction_x0[genes[1]]}"
    ax2.text(0.7, 0.4, relationship_count_text, fontsize=14)
    


```

```python
plot_diversity_comparison('oc43', ['s1', 's2'])
```
