```python
#Find ancestral genome seq for OC43 lineage A
genome_root_file = '../oc43/auspice/seasonal_corona_oc43_full_root-sequence.json'
with open(genome_root_file, "r") as root_handle:
    genome_root = json.load(root_handle)
    genome_root_seq = genome_root['nuc']

lineage_a_root_genome = str(genome_root_seq)  


for mut in a_muts_from_root_genome:
    pos = int(mut[1:-1])-1
    new_nt = mut[-1]
    lineage_a_root_genome = lineage_a_root_genome[:pos]+new_nt+lineage_a_root_genome[pos+1:]

print(len(lineage_a_root_genome))
print(lineage_a_root_genome)
```

```python
#whole genome empirical data for oc43a lineage
lineage_a = []
lineage_b = []
lineage_file = '../oc43/results/clades_full.json'

with open(lineage_file, "r") as lineage_handle:
    lineages = json.load(lineage_handle)
    for node, v in lineages['nodes'].items():
        if 'NODE' not in node:
            if v['clade_membership'] =='A':
                lineage_a.append(node)
            elif v['clade_membership'] == 'B':
                lineage_b.append(node)
```

```python
#whole genome empirical data for oc43a lineage
lineage_a_records = []
lineage_b_records = []
with open('../oc43/data/oc43_full.fasta', "r") as handle:
    for virus in SeqIO.parse(handle, "fasta"):
        virus_id = virus.id.split('|')[0]
        if virus_id in lineage_a:
            if virus_id == 'oc43/KF963229/HCoV_OC43/FRA_EPI/Caen/1967/VR759/1967':
                pass
            else:
                new_record_description = virus_id + '|'+ virus.id.split('|')[3]
                lineage_a_records.append(SeqRecord(virus.seq, id=new_record_description, description=new_record_description))
        if virus_id in lineage_b:
            new_record_description = virus_id + '|'+ virus.id.split('|')[3]
            lineage_b_records.append(SeqRecord(virus.seq, id=new_record_description, description=new_record_description))
            
SeqIO.write(lineage_a_records, 'oc43a_empirical_genome.fasta', "fasta")
SeqIO.write(lineage_b_records, 'oc43b_empirical_genome.fasta', "fasta")
            
```

```python
#spike only
clade_a = []
clade_b = []
clade_file = '../oc43/results/clades_spike.json'

with open(clade_file, "r") as clade_handle:
    clades = json.load(clade_handle)
    for node, v in clades['nodes'].items():
        if 'NODE' not in node:
            if v['clade_membership'] =='A':
                clade_a.append(node)
            elif v['clade_membership'] == 'B':
                clade_b.append(node)
```
